```python
import pandas as pd
from io import StringIO
```

```python
y_start = 1980
y_end = 2023

folder = 'earthquake_catalog/'
ext = '.catalog'
for y in range(y_start, y_end+1):
    #print(y)
    file = folder + str(y) + ext
    with open(file, 'r') as f:
        lines = f.readlines()
    lines = lines[9:]
    text = '\n'.join(lines)
    text = text.replace('  ',' ')
    text = text.replace('  ',' ')
    text = text.replace('  ',' ')
    if y == y_start:
        df = pd.read_csv(StringIO(text),delimiter=' ')
    else:
        df = df.append(pd.read_csv(StringIO(text),delimiter=' '))      
df.head(-10)
```

```python
df.to_csv('earthquake_catalog_1980_2023.csv')
```
